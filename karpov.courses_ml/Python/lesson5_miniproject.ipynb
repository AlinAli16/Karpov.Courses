Описание данных

App — название приложения

Category — категория, к которой относится приложение

Rating — рейтинг пользователей

Reviews — количество отзывов пользователей о приложении

Size — размер приложения

Installs — количество загрузок/установок приложения пользователями

Type — платное или бесплатное приложение

Price — цена приложения

Content Rating — возрастная группа, на которую ориентировано приложение

Genres — принадлежность приложения к нескольким жанрам

Last Updated — дата последнего обновления приложения в Play Store

Current Ver — текущая версия приложения в Play Store

Android Ver — минимальная требуемая версия Android

```python
'''
Допустим, мы хотим собрать датафрейм, который будет включать в себя: 
первые три строки исходного датафрейма, 
строки 6-8 (включительно), 
строки 16-19 (включительно) 
и колонки: App, Size, Genres, Current Ver в соответствующем порядке.

По сути, это будет выглядеть как объединение трех, состоящих из нескольких строк, датафреймов, являющимися срезами из исходных данных.

В качестве ответа загрузите полученный датафрейм в формате csv. При сохранении финального датафрейма, колонка с индексами должна остаться.

NB! Помните про индексацию с нуля и правила включения/невключения границ при слайсинге
'''

# создадим 3 среза
slice1 = playstore.loc[:2] # первые три строки
slice2 = playstore.loc[5:7] # строки 6-8
slice3 = playstore.loc[15:18] # строки 16-19

# оставим только нужные колонки в каждом срезе
slice1 = slice1[['App', 'Size', 'Genres', 'Current Ver']]
slice2 = slice2[['App', 'Size', 'Genres', 'Current Ver']]
slice3 = slice3[['App', 'Size', 'Genres', 'Current Ver']]

# конкатенируем три среза в один датафрейм
result = pd.concat([slice1, slice2, slice3], axis=0)

# сохраняем результат
#result.to_csv('result_concat.csv')
```

```python
'''
Для дальнейшего анализа нам необходимо убрать дубликаты приложений (колонка App) и сбросить индекс. 
Сохраните полученный результат в виде датафрейма в переменную unique_playstore. 
'''
unique_playstore = playstore.drop_duplicates(subset='App').reset_index(drop=True)
unique_playstore
```

```python
'''
В этом задании названия колонок необходимо привести к стандартному виду — 
все буквы должны быть нижнего регистра, 
а пробелы должны быть заменены на нижние подчеркивания. 
'''
playstore.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)
playstore.head()
```

```python
'''
Давайте посмотрим на долю платных и бесплатных приложений. 
В качестве ответа введите долю бесплатных приложений. 
Ответ округлите до сотых, в качестве десятичного разделителя используйте точку.

NB! Не забудьте убрать дубликаты приложений (колонка App) перед подсчетом доли!
'''
#unique_playstore = playstore.drop_duplicates(subset='App').reset_index(drop=True)
unique_playstore.Type.value_counts(normalize=True).round(2)
```

```python
'''
Отберите из датафрейма только те приложения, которые относятся к образовательной категории  (EDUCATION) 
и с количеством отзывов пользователей более 1000. Сохраните новый датафрейм в переменную education_playstore.

Данные очищены от дубликатов, названия колонок преобразованы согласно предыдущим заданиям и сохранены в датафрейме playstore.

NB! Не забудьте сбросить индекс в полученном датафрейме!
'''
# unique_playstore - датафрейм без дубликатов, подготовленный выше
# приведем к стандартному виду дата без дубликатов
unique_playstore.rename(columns=lambda x: x.strip().lower().replace(" ", "_"), inplace=True)

education_playstore = unique_playstore[(unique_playstore.category == "EDUCATION") & (unique_playstore.reviews > 1000)].reset_index(drop=True)
education_playstore
```

```python
'''
Давайте уберем лишние символы из колонки с ценой (price), чтобы далее с ней было удобнее работать, и переведем данные в тип float. 
Данные уже сохранены в playstore.
'''
playstore['price'] = playstore.price.apply(lambda x: x.replace('$', ''))
playstore['price'] = playstore.price.astype(float)
```

```python
# Предыдущий шаг, но для таблица без дубликатов
unique_playstore['price'] = unique_playstore.price.apply(lambda x: x.replace('$', ''))
unique_playstore['price'] = unique_playstore.price.astype(float)
```

```python
'''
В качестве последнего задания вам предстоит сделать сводную таблицу по данным. 
Полученная на выходе таблица должна соответствовать представленному шаблону: 

Колонки должны иметь соответствующие названия, значения таблицы округлены до указанного числа знаков после точки. 
В качестве ответа загрузите полученную таблицу в формате csv (сепаратор запятая - ,).

NB! Понадобится погуглить информацию по запросу "Сводная таблица pandas"

NB! Таблицу нужно делать по данным, где удалены дубликаты (unique_playstore)
'''
pivot_playstore = pd.pivot_table(data=unique_playstore, values = ['price','rating','reviews'], index = ['category','type'], 
                                 aggfunc="mean")
pivot_playstore = pivot_playstore.rename(columns={'price': 'mean_price', 'rating': 'mean_rating', 'reviews': 'mean_reviews'})

pivot_playstore['mean_price'] = pivot_playstore.mean_price.round(2)
pivot_playstore['mean_rating'] = pivot_playstore.mean_rating.round(1)
pivot_playstore['mean_reviews'] = pivot_playstore.mean_reviews.round(2)

pivot_playstore

#pivot_playstore.to_csv('pivot_playstore.csv', sep=',')
```
